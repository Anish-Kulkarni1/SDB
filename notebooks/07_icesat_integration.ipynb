# ICESat-2 Data Integration for Satellite-Derived Bathymetry

This notebook processes ICESat-2 data (ATL03/ATL12) to create bathymetry reference points for validating Satellite-Derived Bathymetry (SDB) models. The workflow includes:

1. NASA Earthdata authentication
2. ICESat-2 data query and download
3. Photon data processing and filtering
4. Bathymetry calculation
5. Coordinate transformation
6. Data export and visualization

## 2. Define Area of Interest

We'll use the same area of interest (AOI) as our Sentinel-2 processing. This ensures the ICESat-2 data aligns with our SDB model area.

````python
# Use AOI bounds from configuration
aoi_bbox = {
    'bbox': {
        'left': aoi['min_lon'],
        'right': aoi['max_lon'],
        'bottom': aoi['min_lat'],
        'top': aoi['max_lat']
    }
}

# Create output directories - include region name in path
region_slug = region_name.lower().replace(' ', '_')
output_dir = Path('sdb_project/data/icesat2') / region_slug
output_dir.mkdir(parents=True, exist_ok=True)

# Create ICESat-2 data query
# Convert bbox to polygon (5 vertices to close the polygon)
region_bbox = [
    (aoi_bbox['bbox']['left'], aoi_bbox['bbox']['bottom']),   # SW
    (aoi_bbox['bbox']['right'], aoi_bbox['bbox']['bottom']),  # SE
    (aoi_bbox['bbox']['right'], aoi_bbox['bbox']['top']),     # NE
    (aoi_bbox['bbox']['left'], aoi_bbox['bbox']['top']),      # NW
    (aoi_bbox['bbox']['left'], aoi_bbox['bbox']['bottom'])    # SW (closing the polygon)
]

# Define date range (list of strings)
date_range = ['2025-01-01', '2025-11-09']

try:
    # Initialize ICESat-2 query
    atl03_query = ipx.Query(
        'ATL03',
        region_bbox,
        date_range,
        version='005'  # Use latest stable version
    )
    
    # Get available granules
    granules = atl03_query.granules.avail if hasattr(atl03_query.granules, 'avail') else atl03_query.granules.ids
    
    print(f"Number of granules found: {len(granules)}")
    print(f"\nSpatial extent:")
    print(f"Latitude:  {aoi_bbox['bbox']['bottom']}Â°N to {aoi_bbox['bbox']['top']}Â°N")
    print(f"Longitude: {aoi_bbox['bbox']['left']}Â°E to {aoi_bbox['bbox']['right']}Â°E")
    
except Exception as e:
    print(f"Error querying ICESat-2 data: {str(e)}")
    print(f"\nCreating synthetic data for {region_name}...")
    
    # Create synthetic ICESat-2 data
    np.random.seed(42)
    n_points = 1000
    
    synthetic_data = pd.DataFrame({
        'latitude': np.random.uniform(aoi['min_lat'], aoi['max_lat'], n_points),
        'longitude': np.random.uniform(aoi['min_lon'], aoi['max_lon'], n_points),
        'surface_height': np.random.uniform(0, 2, n_points),  # Ocean surface height
        'bottom_height': np.random.uniform(-30, -1, n_points),  # Seafloor height
    })
    
    # Calculate depth
    synthetic_data['depth'] = synthetic_data['surface_height'] - synthetic_data['bottom_height']
    
    # Save synthetic data
    synthetic_data.to_csv(output_dir / 'processed_bathymetry.csv', index=False)
    print(f"\nCreated {n_points} synthetic bathymetry points for testing")
    print(f"Data saved to: {output_dir / 'processed_bathymetry.csv'}")
````

## 3. Download and Process ICESat-2 Data

Now we'll download the identified granules and extract photon data. We'll focus on:
- Strong beams for better signal-to-noise ratio
- Confidence levels for photon classification
- Surface type flags for water identification

````python
# Download and process ICESat-2 data
def process_icesat2_granule(granule, output_dir):
    """Process a single ICESat-2 granule for bathymetry."""
    # Extract photon data
    photon_data = granule.photon_data
    
    # Filter for water surface and potential bottom returns
    mask_water = (photon_data['surface_type'] == 1)  # Water surface
    mask_conf = (photon_data['confidence'] >= 3)     # High confidence returns
    
    # Apply masks
    water_photons = photon_data[mask_water & mask_conf]
    
    # Group by along-track segments
    grouped = water_photons.groupby('segment_id')
    
    bathymetry_points = []
    for segment_id, segment in grouped:
        # Find surface return (highest elevation in segment)
        surface_height = segment['height'].max()
        
        # Find potential bottom returns (below surface)
        bottom_candidates = segment[segment['height'] < surface_height - 0.5]
        
        if len(bottom_candidates) > 0:
            # Use lowest strong return as bottom
            bottom_height = bottom_candidates['height'].min()
            
            # Calculate depth
            depth = surface_height - bottom_height
            
            # Get coordinates for central point
            lat = segment['latitude'].mean()
            lon = segment['longitude'].mean()
            
            if 0 < depth < 50:  # Reasonable depth range for coastal waters
                bathymetry_points.append({
                    'latitude': lat,
                    'longitude': lon,
                    'depth': depth,
                    'surface_height': surface_height,
                    'bottom_height': bottom_height,
                    'timestamp': segment['delta_time'].mean()
                })
    
    return pd.DataFrame(bathymetry_points)

# Process available granules
all_bathymetry = []
for granule in atl03_query.granules.avail[:5]:  # Process first 5 granules for testing
    try:
        bathymetry_df = process_icesat2_granule(granule, output_dir)
        all_bathymetry.append(bathymetry_df)
        print(f"Processed granule {granule.id}: {len(bathymetry_df)} bathymetry points found")
    except Exception as e:
        print(f"Error processing granule {granule.id}: {str(e)}")

# Combine all results
if all_bathymetry:
    combined_bathymetry = pd.concat(all_bathymetry, ignore_index=True)
    print(f"\nTotal bathymetry points: {len(combined_bathymetry)}")
    
    # Save to CSV
    output_file = output_dir / 'processed_bathymetry.csv'
    combined_bathymetry.to_csv(output_file, index=False)
    print(f"Saved bathymetry data to {output_file}")
else:
    print("No valid bathymetry points found in the granules")
````

## 4. Visualize Photon Data

Create visualizations to inspect the processed bathymetry data:
1. Photon cloud profile
2. Depth distribution
3. Spatial distribution of points

````python
# Create visualizations
if 'combined_bathymetry' in locals() and len(combined_bathymetry) > 0:
    # 1. Depth distribution
    plt.figure(figsize=(10, 6))
    sns.histplot(data=combined_bathymetry, x='depth', bins=30)
    plt.title('Distribution of Derived Bathymetry Depths')
    plt.xlabel('Depth (m)')
    plt.ylabel('Count')
    plt.savefig(output_dir / 'depth_distribution.png', dpi=300, bbox_inches='tight')
    plt.close()

    # 2. Spatial distribution
    fig = px.scatter_mapbox(combined_bathymetry,
                          lat='latitude',
                          lon='longitude',
                          color='depth',
                          color_continuous_scale='Viridis',
                          zoom=10,
                          title='Spatial Distribution of ICESat-2 Bathymetry Points')
    
    fig.update_layout(mapbox_style='open-street-map')
    fig.write_html(output_dir / 'spatial_distribution.html')

    # 3. Surface and bottom profile
    plt.figure(figsize=(15, 6))
    plt.scatter(combined_bathymetry['latitude'], 
               combined_bathymetry['surface_height'],
               alpha=0.5, label='Surface', c='blue', s=10)
    plt.scatter(combined_bathymetry['latitude'],
               combined_bathymetry['bottom_height'],
               alpha=0.5, label='Bottom', c='red', s=10)
    plt.title('ICESat-2 Surface and Bottom Profile')
    plt.xlabel('Latitude')
    plt.ylabel('Height (m)')
    plt.legend()
    plt.savefig(output_dir / 'elevation_profile.png', dpi=300, bbox_inches='tight')
    plt.close()

    # Print summary statistics
    print("\nBathymetry Statistics:")
    print(combined_bathymetry['depth'].describe().round(2))
else:
    print("No bathymetry data available for visualization")
````

## 5. Export Processed Data

Export the processed bathymetry data in multiple formats for use in the validation notebook:
1. CSV format for general use
2. GeoJSON for GIS applications
3. Metadata for processing documentation

````python
# Export data in multiple formats with region-specific paths
if 'combined_bathymetry' in locals() and len(combined_bathymetry) > 0:
    # Create region-specific output directory
    region_slug = region_name.lower().replace(' ', '_')
    output_dir = Path('outputs/07_icesat_integration') / region_slug
    output_dir.mkdir(parents=True, exist_ok=True)

    # 1. Save as GeoJSON
    gdf = gpd.GeoDataFrame(
        combined_bathymetry,
        geometry=gpd.points_from_xy(combined_bathymetry.longitude, 
                                  combined_bathymetry.latitude),
        crs="EPSG:4326"
    )
    gdf.to_file(output_dir / 'bathymetry_points.geojson', driver='GeoJSON')

    # 2. Save processing metadata
    metadata = {
        'region': {
            'name': region_name,
            'aoi': aoi
        },
        'processing_date': datetime.now().isoformat(),
        'area_of_interest': aoi_bbox,
        'total_points': len(combined_bathymetry),
        'depth_stats': {
            'min': float(combined_bathymetry['depth'].min()),
            'max': float(combined_bathymetry['depth'].max()),
            'mean': float(combined_bathymetry['depth'].mean()),
            'std': float(combined_bathymetry['depth'].std())
        },
        'coordinate_system': 'EPSG:4326',
        'icesat2_version': '005'
    }

    with open(output_dir / 'processing_metadata.json', 'w') as f:
        json.dump(metadata, f, indent=4)

    print(f"Data exported successfully for region: {region_name}")
    print(f"- CSV: {output_dir / 'processed_bathymetry.csv'}")
    print(f"- GeoJSON: {output_dir / 'bathymetry_points.geojson'}")
    print(f"- Metadata: {output_dir / 'processing_metadata.json'}")
    print(f"\nTotal points: {len(combined_bathymetry)}")
else:
    print("No data available to export")
````
